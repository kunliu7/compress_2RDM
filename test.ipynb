```python
import numpy as np
```

```python
n = 6
a = np.array(range(n ** 4)).reshape((n,) * 4)
print(a)
```

```python

b = a.reshape((n * n,) * 2)
print(b)
```

```python
c = a.transpose([1, 0, 2, 3])
print(c)
```

```python
d = c.reshape((n * n,) * 2)
print(d)
```

```python
print(a[1, 0, 1, 1])
print(c[0, 1, 1, 1])
print(np.argwhere(b==223))
print(np.argwhere(d==223))
```
